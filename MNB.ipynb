```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def load_data(folder_path):
    reviews = []
    labels = []
    for label in ['pos', 'neg']:
        dir_path = os.path.join(folder_path, label)
        for filename in os.listdir(dir_path):
            filepath = os.path.join(dir_path, filename)
            with open(filepath, 'r', encoding='utf-8') as f:
                reviews.append(f.read())
            labels.append(1 if label == 'pos' else 0)

    df = pd.DataFrame({'review': reviews, 'label': labels})
    return df

train_df = load_data('train')
test_df  = load_data('test')

train_df.head()
```

Text Preprocessing

```python
import string

def preprocess_text(text):
    text = text.lower()
    text = text.replace("<br />", " ")
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    return tokens

train_df['tokens'] = train_df['review'].apply(preprocess_text)
test_df['tokens'] = test_df['review'].apply(preprocess_text)

train_df.head()
```

Build Vocabulary

```python
def build_vocabulary(tokens_series):
    vocab = {}
    index = 0
    for tokens in tokens_series:
        for word in tokens:
            if word not in vocab:
                vocab[word] = index
                index += 1
    return vocab

vocab = build_vocabulary(train_df['tokens'])
print(f"Vocabulary size: {len(vocab)}")
```

Bag of Words

```python
def bag_of_words(tokens, vocab):
    bow = [0] * len(vocab)
    for word in tokens:
        if word in vocab:
            bow[vocab[word]] += 1
    return bow

# apply to both dataframes using the vocab built from training
train_df['bow'] = train_df['tokens'].apply(lambda tokens: bag_of_words(tokens, vocab))
test_df['bow']  = test_df['tokens'].apply(lambda tokens: bag_of_words(tokens, vocab))

train_df.head()
```
